```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('Genetic11-11-2025 13-14.csv')

# Normalize delay to 0..1 (0 = best)
dmax = df["BestDelay"].max()
colors = 1 - (df["BestDelay"] / dmax)

plt.figure(figsize=(6,5))
plt.scatter(df["Crossover"], df["Mutation"], s=100, c=colors, cmap='Greens')
plt.xlabel("Crossover Probability")
plt.ylabel("Mutation Probability")
plt.title("Correlation of Mutation vs Crossover\nColor = closeness to Delay=0")
plt.colorbar(label="Green intensity")
plt.tight_layout()

plt.show()
```
